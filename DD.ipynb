```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the Zachary Karate Club graph
# G = nx.karate_club_graph()
G = nx.les_miserables_graph()
# G = nx.gnp_random_graph(50, 0.5, directed=True)

# Calculate the degree distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_counts = nx.degree_histogram(G)

#Plot the graph
pos = nx.spring_layout(G)  # Positions for all nodes
plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
plt.title("Les miserables graph")
plt.show()


# Plot the degree distribution
# plt.figure(figsize=(10, 10))
# plt.bar(range(len(degree_counts)), degree_counts, width=0.8, color='b')
# plt.title("Degree Distribution of Les miserables graph")
# plt.xlabel("Degree")
# plt.ylabel("Frequency")
# plt.xticks(range(len(degree_counts)), range(len(degree_counts)))
# plt.show()
```

```python
# Calculate the probabilities
total_nodes = G.number_of_nodes()
degree_probabilities = [count / total_nodes for count in degree_counts]

# Plot the degree distribution
plt.figure(figsize=(10, 10))
plt.bar(range(len(degree_probabilities)), degree_probabilities, width=0.8, color='b')
plt.title("Degree Distribution Probability of Les miserables graph")
plt.xlabel("Degree")
plt.ylabel("Probability")
plt.xticks(range(len(degree_counts)), range(len(degree_counts)))
plt.show()
```
